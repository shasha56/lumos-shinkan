## 今回のchat botを作るうえで必要な構文

### 1. `input()`：入力を受け取る
`input()` は、キーボードから打った文字をプログラムの中に取り込む箱
- 例： `name = input("名前は？")`
- これで、入力した名前が `name` という変数（箱）に保存される

```python
name = input("名前は？ ")
print("こんにちは、" + name + "さん！")
```

### 2. `if`：もしも〜なら（条件分岐）
`if`条件によってプログラムの動きを変える
- 「もしも、入力された言葉が『占い』なら、おみくじを引く」
- 「そうでなければ、挨拶を返す」
みたいな使い分けができる

**⚠️ 注意**

Pythonでは、`if` の中身を書くときに **「半角スペース4つ分」** のスキマを空けるルールがある。これがズレるとプログラムが動かなくなる

```python
fruit = "りんご"
if fruit == "りんご":
    print("これはりんごです。")
elif fruit == "みかん":
    print("これはみかんです。")
else:
    print("これはりんごではありません。")
```

```python
answer = input("今日は元気ですか？ (はい / いいえ): ")

if answer == "はい":
    print("よかったです！")
else:
    print("無理しないでくださいね。")
```

### チャットボットの作成！

```python
# --- 手順1: 名前を入力してもらう ---
# input() を使ってユーザーからの入力を受け取る
user_name = input("君の名前を教えて！ : ")

# --- 手順2: 挨拶を返す ---
# 文字列の前に f をつけると、変数（user_name）を中に埋め込める
print(f"よろしくね、{user_name}くん！")

# --- 手順3: ボットの最初のセリフ
# 自由に書き換えてみよう！例に入力してほしいテーマを書いてみるといいかも。
print("ボット：今日は何についてお話しする？（例：占い、計算、挨拶）")

# --- 手順4: ユーザーの入力に応じて返事をする ---
message = input("あなた：")

# --- 手順5: 条件分岐を使ってテーマごとに返事を変える ---
if message == "占い":
    print("ボット：今日の運勢は『吉』！ラッキーカラーは青だよ。")
elif message == "計算":
    a = int(input("1つ目の数字："))
    b = int(input("2つ目の数字："))
    print(f"ボット：合計は {a + b} です。")
elif message == "挨拶":
    print(f"ボット：やあ、{user_name}さん！今日も来てくれてありがとう。")
else:
    print("ボット：ごめんね、そのテーマはまだ勉強中。")
```

### 余裕のある人は次の構文もチャットボットに組み込んでみよう！

### 3. `while`：くり返し処理

`while` は、**条件がTrue（正しい）のあいだ** 同じ処理を何度も実行する構文。

- 例：`count` が 5 未満のあいだ、数字を表示し続ける
- 条件が False になると、ループは終了

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

### 4. `break`：くり返しを途中で止める

`break` は、`while` の条件に関係なく、**その場でループを終了** させる命令。  
「特定の入力が来たら終わりにしたい」ときによく使います。

```python
while True:
    text = input("終了するには end と入力: ")
    if text == "end":
        print("ループを終了します。")
        break
    print("入力された文字:", text)
```

```python
user_name = input("君の名前を教えて！: ")

print(f"ボット：こんにちは、{user_name}さん！")
print("ボット：話しかけてね。（終了したいときは「おわり」）")

while True:
    message = input("あなた：")

    if message == "おわり":
        print("ボット：またね！")
        break
    else:
        print("ボット：何を言ってるかわからないよ")
```
