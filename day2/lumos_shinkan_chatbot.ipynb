## 今回のchat botを作るうえで必要な構文

### 1. `input()`：入力を受け取る
`input()` は、キーボードから打った文字をプログラムの中に取り込む箱
- 例： `name = input("名前は？")`
- これで、入力した名前が `name` という変数（箱）に保存される

### 2. `if`：もしも〜なら（条件分岐）
`if`条件によってプログラムの動きを変える
- 「もしも、入力された言葉が『占い』なら、おみくじを引く」
- 「そうでなければ、挨拶を返す」
みたいな使い分けができる

**⚠️ 注意**

Pythonでは、`if` の中身を書くときに **「半角スペース4つ分」** のスキマを空けるルールがある。これがズレるとプログラムが動かなくなる

```python
# --- 手順1: 名前を入力してもらう ---
# input() を使ってユーザーからの入力を受け取る
user_name = input("君の名前を教えて！ : ")

# --- 手順2: 挨拶を返す ---
# 文字列の前に f をつけると、変数（user_name）を中に埋め込める
print(f"よろしくね、{user_name}くん！")

# --- 手順3: ボットの性格 ---
# ボットの最初のセリフを自由に書き換えてみる
print("ボット：今日は何についてお話しする？（例：占い、計算、サークル）")
```

```python
import random # ランダムな機能を使うための準備

# ユーザーからの入力を受け取る
user_msg = input(f"{user_name} : ")

# --- ボットの思考ロジック ---
# 条件分岐 if を使って、ユーザーのメッセージに応じた返しをする

if "占い" in user_msg:
    fortunes = ["超大吉", "吉", "末吉", "再履修確定"]
    result = random.choice(fortunes) # リストからランダムに1つ選ぶ
    print(f"ボット：今日の運勢は...【{result}】だよ！")

elif "計算" in user_msg:
    print("ボット：1から100まで全部足すと、いくつになるか知ってる？")
    n = 100
    total = n * (n + 1) // 2
    print(f"ボット：答えは {total} だよ。簡単だね！")

elif "サークル" in user_msg:
    print("ボット：サークルといえば、Lumosというプログラミングサークルがあるよ！")

else:
    # どのキーワードにも当てはまらない時の返し
    print("ボット：すみません。よくわかりません")
```
